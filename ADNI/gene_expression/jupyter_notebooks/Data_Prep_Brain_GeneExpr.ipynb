This notebook is for merging the ADNI Gene Expression Data with The brain structural features as in Xin's paper.

```python
import os
import pandas as pd

#os.listdir()
```

```python
Adni_Merge = pd.read_csv('ADNIMERGE.csv')
Gene_Expr = pd.read_csv('ADNI_Gene_Expression_Profile.csv',low_memory=False)
Xin_Features = pd.read_csv('Combined_Features.csv',encoding= 'unicode_escape')
```

```python
gene_expr_subjects = list(Gene_Expr.iloc[1][3:-1])
```

```python
gene_expr_RID = [Adni_Merge[Adni_Merge['PTID'] == s]['RID'].iloc[0] for s in gene_expr_subjects]
```

```python
len(gene_expr_RID)
```

```python
Xin_RID = list(Xin_Features.RID)
```

```python
common_RID = list(set(Xin_RID) & set(gene_expr_RID)) 
```

```python
len(common_RID)
```

```python
common_Xin = Xin_Features[pd.DataFrame(Xin_Features.RID.tolist()).isin(common_RID).any(1).values]
```

```python
from collections import Counter
Counter(common_Xin.DX)
```

```python
#0=CN, 2=LMCI, 3=AD
```

```python
Counter(Xin_Features.DX)#Original Xin's Feature
```

```python
gene = Gene_Expr.T
gene.columns = gene.iloc[0]
gene
```

```python
gene = gene.drop(['Phase', 'Unnamed: 1','Unnamed: 2','Unnamed: 747'])
gene
```

```python
gene['RID'] = gene_expr_RID
gene
```

```python
gene_brain_common = gene[pd.DataFrame(gene.RID.tolist()).isin(common_RID).any(1).values]
```

```python
gene_brain_final = gene_brain_common.merge(common_Xin, how='inner', on='RID')
```

```python
gene_brain_final.to_csv('GeneExpr_brain_feat.csv')
```

```python
gene_brain_final.head()
```
